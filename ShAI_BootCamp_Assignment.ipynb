```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
# Load your dataset
df = pd.read_csv('Salaries.csv')
df.head()
```

```python
copyData = df.copy()
copyData
```

```python
copyData.info()
```

```python
copyData.rename(columns = {'TotalPayBenefits':'TotalIncome'}, inplace = True)
copyData.rename(columns = {'TotalPay':'Salary'}, inplace = True)
copyData.head()
```

```python
copyData['Department'] = copyData['JobTitle'].str.extract(r'\((.*?)\)')

# Display the updated DataFrame
copyData['Department']
```

```python
plt.figure(figsize=(8, 6))

sns.histplot(copyData.Salary, bins=30)
plt.ylabel('Frequency')
plt.xlabel('Salary')


plt.show()
```

```python
Count_Employee = copyData.shape[0]
Department_Counts = copyData['Department'].value_counts()
plt.figure(figsize=(8, 8))
plt.figure(figsize=(8, 8))
plt.pie(Department_Counts, labels=Department_Counts.index, autopct='%1.1f%%', colors=['red', 'lightskyblue', 'lightgreen', 'gold', 'brown'])
plt.title('Employee Distribution by Department')
plt.axis('equal')
try:
    plt.show()
except TypeError:
    plt.show().item()
```

```python
Grouped_Data = copyData.groupby(['Department','JobTitle']).agg({'Salary': 'mean'}).reset_index()
Grouped_Data
```

```python
Correlation_1 = copyData['Salary'].corr(copyData['OvertimePay'])
print(f"Correlation between Salary and OvertimePay: {Correlation_1}")
plt.scatter(copyData['OvertimePay'], copyData['Salary'])
plt.xlabel('OvertimePay')
plt.ylabel('Salary')
plt.show()
```
